```python
%pip install pandas
import pandas as pd
import os

# get all file names in MedQuAD directory

foldername = [filename for filename in os.listdir()]

# Filter out files that don't start with a digit
foldername = [filename for filename in foldername if filename[0].isdigit()]

# Sort the list of files by the initial number
foldername = sorted(foldername, key=lambda x: int(''.join(filter(str.isdigit, x.split()[0]))))

print(foldername)
```

```python
import xml.etree.ElementTree as ET

# Initialize an empty list to store the data
data = []

# Loop through all files in the directory
for directory in foldername:
    for filename in os.listdir(directory):
        print(filename)
        if filename.endswith(".xml"):
            # Parse the XML file
            tree = ET.parse(os.path.join(directory, filename))
            print(tree)
            root = tree.getroot()
            
            # Find all QAPairs in the XML file
            for qa_pair in root.findall(".//QAPair"):
                question = qa_pair.find('Question').text
                answer = qa_pair.find('Answer').text
                # Append the question and answer to the data list
                data.append({"Question": question, "Answer": answer})
```

```python
# save into csv
df = pd.DataFrame(data)
df.to_csv("MedQuAD-QA-pair.csv", index=False)
```
